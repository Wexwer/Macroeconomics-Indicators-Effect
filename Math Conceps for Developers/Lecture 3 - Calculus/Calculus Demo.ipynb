# Calculus

## Demos

```python
def f(x):
    return x ** 3 - 2 * x + 5
```

```python
x_central = 1
delta= 0.1
x_min, x_max = x_central - delta, x_central + delta
y_min, y_max = f(x_min), f(x_max)
print(f"If x: {x_min, x_max}")
print(f"then y: {y_min, y_max}")
```

```python
def calculate_derivative(f,x_0, tolerance = 0.00001):
   return (f(x_0 + tolerance) - f(x_0)) / tolerance
```

```python
calculate_derivative(lambda x: 3 * x ** 2 + 5 * x - 8, 3)
```
